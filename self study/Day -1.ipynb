<h3> Task 6 </h3>

```python
from math import sin, cos, pi,sqrt

M1 = np.array([
    [1/sqrt(2),1/sqrt(2)],
    [1/sqrt(2),-1/sqrt(2)]
])

M2 = np.array([
    [3,0],
    [0,3]
])

def operator(theta):
    R = np.array([
        [cos(theta),-sin(theta)],
        [sin(theta),cos(theta)]
    ])
    return R
```

```python
def unitary_check(mat):
    matrix_dagger = mat.T.conj()
    res_mul = mat @ matrix_dagger
    identity = np.eye(len(mat))
    are_equal = np.allclose(identity,res_mul)
    return are_equal

    
```

```python
unitary_check(M1)
```

```python
unitary_check(M2)
```

```python
thetas = [pi,pi/2,pi/3]
for theta in thetas:
    opr = operator(theta)
    res = unitary_check(opr)
    print(res)
```

<h3> Task 13 </h3>

For a system with 4 states, randomly create a probabilistic state, and print its entries, e.g., $ 0.16,  0.17,   0.02,  0.65 $.

<i>Hint: You may pick your random numbers between 0 and 100 (or 1000), and then normalize each value by dividing the summation of all numbers.</i>

```python
from numpy import random
rng = random.default_rng(12)
coefficient = rng.integers(0,100,endpoint=True,size=4)
coefficient_normalized = coefficient/np.sum(coefficient)
print(coefficient_normalized)
```
